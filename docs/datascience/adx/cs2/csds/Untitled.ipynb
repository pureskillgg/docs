```python
from io import BytesIO
from gzip import GzipFile
import json
import gzip
```

```python
with gzip.open('csds', 'r') as fin:
    data = json.loads(fin.read().decode('utf-8'))
```

```python
channels = data['channels']
```

```python
def print_row(row):
    data=['']
    data.append(row['name'])
    data.append(row['origin'])
    data.append(', '.join(row.get('dependents',[])))
    data.append(', '.join(row.get('mergeKeys',[])))
    data.append('')
    print('|'.join(data))
    
```

```python
for channel in channels:
    print("## "+channel['channel']+' - ')
    print()
    for column in channel['columns']:
        print_row(column)
    print()
```
